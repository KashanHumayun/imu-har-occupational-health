# IMU HAR Demo Walkthrough

This notebook reads the generated demo artifacts, inspects model performance, and previews per-window predictions.

```python
from pathlib import Path
import json
import pandas as pd

ROOT = Path.cwd().resolve().parents[0] if Path.cwd().name == 'notebooks' else Path.cwd()
REPORTS = ROOT / 'reports' / 'demo'
MODELS = ROOT / 'models' / 'demo'

metrics = json.loads((REPORTS / 'metrics.json').read_text())
preds = pd.read_csv(REPORTS / 'window_predictions.csv')
metrics['best_model'], MODELS
```

```python
pd.DataFrame({
    'model': ['random_forest', 'svm', 'workflow_hmm'],
    'accuracy': [
        metrics['classical_models']['random_forest']['mean_accuracy'],
        metrics['classical_models']['svm']['mean_accuracy'],
        metrics['workflow_hmm']['decoded_accuracy'],
    ],
    'macro_f1': [
        metrics['classical_models']['random_forest']['mean_macro_f1'],
        metrics['classical_models']['svm']['mean_macro_f1'],
        metrics['workflow_hmm']['decoded_macro_f1'],
    ],
}).sort_values('accuracy', ascending=False)
```
