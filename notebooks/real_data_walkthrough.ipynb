# Real Data Walkthrough

This notebook inspects the real UCI HAR and PAMAP2 outputs saved under `reports/results`.

```python
from pathlib import Path
import json
import pandas as pd

ROOT = Path.cwd().resolve().parents[0] if Path.cwd().name == 'notebooks' else Path.cwd()
REPORTS = ROOT / 'reports' / 'results'
MODELS = ROOT / 'models' / 'results'

metrics = json.loads((REPORTS / 'metrics.json').read_text())
uci_preds = pd.read_csv(REPORTS / 'uci_har_predictions.csv')
pamap_preds = pd.read_csv(REPORTS / 'pamap2_predictions.csv')
metrics['uci_har']['best_model'], metrics['pamap2']['best_model'], MODELS
```

```python
pd.DataFrame({
    'dataset': ['uci_har', 'pamap2'],
    'best_model': [metrics['uci_har']['best_model'], metrics['pamap2']['best_model']],
    'accuracy': [
        metrics['uci_har']['models'][metrics['uci_har']['best_model']]['accuracy'],
        metrics['pamap2']['classical_models'][metrics['pamap2']['best_model']]['mean_accuracy'],
    ],
})
```
